#### 1.1. Load image
Load the example image **exampleImg.jpg**. 
You can use **scikit-image** library to read the image, for example **skimage.io.imread()**. Details please see: http://scikit-image.org/docs/dev/api/skimage.io.html#skimage.io.imread


```python
from skimage import io

# Load exampleImg.jpg, using skimage.io.imread()
exampleImg = io.imread('exampleImg.jpg')
```

#### 1.2. Detect face
Using the dlib to detect face and landmarks on your example image. We provide the basic pipeline here:
1. dlib.get_frontal_face_detector()
2. dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")
Please see http://dlib.net/python/index.html#dlib.get_frontal_face_detector and http://dlib.net/python/index.html#dlib.shape_predictor for more details.

```python
# Loading required libraries
import dlib

# Initializing face detector and shape predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

# Detect face, return rectangles.
dets = detector(exampleImg, 1)

# Return the first rectanges.
shape = predictor(exampleImg, dets[0])
```

#### 1.3. Extract facial landmarks
**We provide function here to help you to transfer the detected shape to landmark positions. Please read the the function and use it.**

```python
import numpy as np
# We provide the shape to points function.
def shape2points(shape, dtype = 'int', pointNum = 68):
    
    coords = np.zeros((pointNum, 2), dtype=dtype)

    for i in range(0, 68):
        coords[i] = (shape.part(i).x, shape.part(i).y)
    
    return coords
```

**Please write your code below to transfer the shape to points:**


```python
# Extract facial landmarks from shape
keypoints = shape2points(shape)
```

#### 1.4. Face normalization
Use **cropFace** to crop the facial image from the background. For details on the **cropFace** function, please read the function in **exercise1Lib.py**. Usually, we have to do the face registration to remove the subject characteristics. To do a  face registration, firstly you need a set of landmarks of a standard face model. Then you calculate the transformation between you detected landmarks and the standard landmarks. Last you apply the transformation on your example image/

**We offer the function to load a standard face model, which contains the mean position of landmarkds estimated from the training data**

```python
import csv
standardModel = np.zeros((68, 2))
tmp = 0 
with open('mean.csv', 'rb') as csvfile:
    spamreader = csv.reader(csvfile, delimiter=',')
    for row in spamreader:
        standardModel[tmp, 0] = float(row[0])
        standardModel[tmp, 1] = float(row[1])
        tmp += 1
standardModel = standardModel * 500
```

**Then you need to calculate the transformation using skimage.transform.PolynomialTransform()**

```python
from skimage import transform
myTransformation = transform.PolynomialTransform()
myTransformation.estimate(standardModel, keypoints)
registeredImg = transform.warp(exampleImg, myTransformation)

```

**Cropping the original example image according to the detected landmarks, and the registered image according to the standard face mode. **

```python
# Crop the face using the provided cropFace function.
from exercise1Lib import cropFace
normedFace = cropFace(exampleImg, keypoints, doNormalization = True)
cropedRegisteredImg = cropFace(registeredImg, standardModel, doNormalization = True)

```

#### 1.5 Show your result
* Plot facial landmarks on the image.
* Plot your cropped image and the corresponding gray histogram.

**Display the original example image, croped example image, registered image, cropped registered image using subplots. You need to place corresponding landmakrs on the two uncorped images.**

You will use Support Vector Machine (SVM) for facial expression classification. 

#### 3.1 Load data
Firstly, you need to read **.mat** files using python. You can use the **scipy.io.loadmat()** function to read .mat file. In the provided **Task3_data.mat** file, different data are packed by different dictionaries which are list below:
* training_data

* training_class
* testing_class
For example if you would like to load the **training_data** from the file, you can use **scipy.io.loadmat('Task3_data.mat')['training_data']**



```python
import sklearn
import scipy.io as sio
# Loading data using scipy.io.loadmat, or sio.loadmat
# Alternatively, you can us h5py also, if you would like to
mdata = sio.loadmat('Task3_data.mat')

sample_train = mdata['training_data']
sample_test = mdata['testing_data']
label_train = mdata['training_class'].reshape(-1)
label_test = mdata['testing_class'].reshape(-1)

```

#### 3.2 Train SVM classifiers
Use the **sklearn.svm** library to train Support Vector Machine (SVM) classifiers. The training_data and testing_data matrices contain the calculated LBP-TOP features for the training and testing sets, respectively. The block size for LBP-TOP used for training and testing data are 2x2x1. The training_class group vector contains the class of samples: 1 = happy, 2 = sadness, corresponding to the rows of the training data matrices.

You need to do:
* Construct an SVM classifier using the linear kernel (e.g. **sklearn.svm.SVC()**). Usage about using the library please see http://scikit-learn.org/stable/modules/generated/sklearn.svm.SVC.html#sklearn.svm.SVC.

* Use the **fit()** function and the **training_data** and **training_class to train your classifier.

* Use your trainined classifer to classify the **training_data** and **testing_data**. Then, calculate average classification performances for both training and testing data. The correct class labels corresponding with the rows of the training and testing data matrices are in the variables **training_class** and **testing_class**, respectively.

* Calculate the average classification performances for the training data (training_data) and the testing data (testing_data) using the corresponding trained linear SVM.

* Calculate confusion matrices for the training and testing data for both classifiers. For example, using the **sklearn.metrics.confusion_matrix()** function.



```python
from sklearn import svm
# Initializing a SVM classifier, using linear kernel
clf = svm.SVC(kernel  = 'linear', cache_size = 5000)

# using the classifier to fit your training data
clf.fit(sample_train, label_train)  


# Predicting you training data and testing data.
prediction_train = clf.predict(sample_train)
prediction = clf.predict(sample_test)

# Calculating the accuracies of your prediction on training data and testing data, respectively.
```

#### 3.2 Show and check the results
* The average classification performances for the training data (**training_data**) and the testing data (**testing_data**) using the corresponding trained linear SVM.
* Calculate the confusion matrix of the linear kernel SVM using the **training_data**
* Calculate the confusion matrix of the  linear kernel SVM using the **testing_data**

```python
print (sum(prediction_train == label_train) * 1.0 / prediction_train.shape[0])
print (sum(prediction == label_test) * 1.0 / prediction.shape[0])


from sklearn.metrics import confusion_matrix

# Draw your confusion matrix
# using sklearn.metrics.confusion_matrix
print (confusion_matrix(label_train, prediction_train))
print (confusion_matrix(label_test, prediction))
```
